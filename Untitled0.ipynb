```python

# Download IMDB Large Movie Review Dataset (Stanford) and convert to CSV
import os, csv
from pathlib import Path

RAW = BASE / "data" / "raw"
ACL_TARBALL = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
ACL_DIR = RAW / "aclImdb"
CSV_OUT = RAW / "reviews.csv"

# Download and extract (skip if present)
if not ACL_DIR.exists():
    %cd {RAW}
    !curl -L -o aclImdb_v1.tar.gz {ACL_TARBALL}
    !tar -xzf aclImdb_v1.tar.gz
    %cd /content

# Combine train/test pos/neg into a single CSV (text,label)
def iter_reviews(split_dir):
    for label_name, label_val in [("pos", 1), ("neg", 0)]:
        directory = split_dir / label_name
        for fname in directory.glob("*.txt"):
            text = fname.read_text(encoding="utf-8", errors="ignore").strip()
            yield text, label_val

rows = []
for split in ["train", "test"]:
    rows.extend(list(iter_reviews(ACL_DIR / split)))

CSV_OUT.parent.mkdir(parents=True, exist_ok=True)
with open(CSV_OUT, "w", encoding="utf-8", newline="") as f:
    w = csv.writer(f)
    w.writerow(["text", "label"])
    for text, label in rows:
        w.writerow([text, label])

print(f"Wrote {len(rows)} rows to {CSV_OUT}")
```

```python

from pathlib import Path

# ---------------- src/config.py ----------------
(Path(BASE / "src" / "config.py")).write_text("""\
RANDOM_STATE = 42
TEST_SIZE = 0.2
VAL_SIZE = 0.1
MAX_FEATURES = 100_000
NGRAM_RANGE = (1, 2)
LOGREG_C = 2.0
LOGREG_MAX_ITER = 1000
""")

# ---------------- src/preprocess.py ----------------
(Path(BASE / "src" / "preprocess.py")).write_text("""\
import re
import pandas as pd
from sklearn.model_selection import train_test_split
from .config import RANDOM_STATE, TEST_SIZE, VAL_SIZE

def clean_text(s: str) -> str:
    s = s.lower()
    s = re.sub(r"[^a-z0-9\\s']", " ", s)
    s = re.sub(r"\\s+", " ", s).strip()
    return s

def load_and_split(path: str, test_size: float = TEST_SIZE, val_size: float = VAL_SIZE):
    df = pd.read_csv(path)
    if 'text' not in df.columns or 'label' not in df.columns:
        raise ValueError("Input CSV must have columns: 'text' and 'label'.")
    df['text'] = df['text'].astype(str).apply(clean_text)
    train_df, test_df = train_test_split(
        df, test_size=test_size, stratify=df['label'], random_state=RANDOM_STATE
    )
    train_df, val_df = train_test_split(
        train_df, test_size=val_size, stratify=train_df['label'], random_state=RANDOM_STATE
    )
    return train_df.reset_index(drop=True), val_df.reset_index(drop=True), test_df.reset_index(drop=True)
""")

# ---------------- src/train_baseline.py ----------------
(Path(BASE / "src" / "train_baseline.py")).write_text("""\
import json, os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from .preprocess import load_and_split
from .config import NGRAM_RANGE, MAX_FEATURES, LOGREG_C, LOGREG_MAX_ITER

def train_baseline(data_path: str,
                   model_out: str = "models/baseline/model.pkl",
                   report_out: str = "reports/metrics_baseline.json"):
    train_df, val_df, test_df = load_and_split(data_path)
    pipe = Pipeline([
        ("tfidf", TfidfVectorizer(ngram_range=NGRAM_RANGE, max_features=MAX_FEATURES)),
        ("clf", LogisticRegression(max_iter=LOGREG_MAX_ITER, C=LOGREG_C, class_weight="balanced"))
    ])
    pipe.fit(train_df['text'], train_df['label'])

    # Validate
    val_pred = pipe.predict(val_df['text'])
    metrics = {
        "val_accuracy": float(accuracy_score(val_df['label'], val_pred)),
        "val_f1": float(f1_score(val_df['label'], val_pred, average="binary")),
        "val_report": classification_report(val_df['label'], val_pred, output_dict=True)
    }

    # Save model & metrics
    import joblib
    os.makedirs(os.path.dirname(model_out), exist_ok=True)
    os.makedirs(os.path.dirname(report_out), exist_ok=True)
    joblib.dump(pipe, model_out)
    with open(report_out, "w") as f:
        json.dump(metrics, f, indent=2)

    # Test
    test_pred = pipe.predict(test_df['text'])
    test_metrics = {
        "test_accuracy": float(accuracy_score(test_df['label'], test_pred)),
        "test_f1": float(f1_score(test_df['label'], test_pred, average="binary")),
        "confusion_matrix": confusion_matrix(test_df['label'], test_pred).tolist()
    }
    os.makedirs("reports", exist_ok=True)
    with open("reports/metrics_baseline_test.json", "w") as f:
        json.dump(test_metrics, f, indent=2)

    return train_df, val_df, test_df, val_pred, test_pred
""")

# ---------------- src/evaluate.py ----------------
(Path(BASE / "src" / "evaluate.py")).write_text("""\
import json
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

def evaluate(y_true, y_pred,
             out_json: str = "reports/metrics.json",
             out_png: str = "reports/confusion_matrix.png"):
    metrics = classification_report(y_true, y_pred, output_dict=True)
    with open(out_json, "w") as f:
        json.dump(metrics, f, indent=2)

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues")
    plt.title("Confusion Matrix (Baseline)")
    plt.savefig(out_png, dpi=200, bbox_inches="tight")
    plt.close()
""")

# ---------------- src/diagnostics.py ----------------
(Path(BASE / "src" / "diagnostics.py")).write_text("""\
import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve

def feature_importance(model_path: str = "models/baseline/model.pkl",
                        out_png: str = "reports/feature_importance.png",
                        top_k: int = 20):
    pipe = joblib.load(model_path)
    tfidf = pipe.named_steps['tfidf']
    clf = pipe.named_steps['clf']
    vocab = np.array(tfidf.get_feature_names_out())
    coefs = clf.coef_[0]

    pos_idx = np.argsort(coefs)[-top_k:][::-1]
    neg_idx = np.argsort(coefs)[:top_k]
    pos_words, pos_vals = vocab[pos_idx], coefs[pos_idx]
    neg_words, neg_vals = vocab[neg_idx], coefs[neg_idx]

    plt.figure(figsize=(10, 6))
    plt.barh(range(top_k), pos_vals[::-1], color='#2ca02c')
    plt.yticks(range(top_k), pos_words[::-1])
    plt.title('Top Positive TFâ€“IDF Features (LogReg Coefficients)')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()

def calibration_plot(model_path: str = "models/baseline/model.pkl",
                     X_texts=None, y_true=None,
                     out_png: str = "reports/calibration_curve.png"):
    if X_texts is None or y_true is None:
        raise ValueError("Provide X_texts and y_true for calibration plot.")
    pipe = joblib.load(model_path)
    probs = pipe.predict_proba(X_texts)[:, 1]
    frac_pos, mean_pred = calibration_curve(y_true, probs, n_bins=10)
    plt.figure(figsize=(6, 6))
    plt.plot(mean_pred, frac_pos, "s-", label="Baseline")
    plt.plot([0,1],[0,1], "k--", label="Perfectly calibrated")
    plt.xlabel("Mean predicted probability")
    plt.ylabel("Fraction of positives")
    plt.title("Calibration Curve")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()
""")

# ---------------- src/predict.py ----------------
(Path(BASE / "src" / "predict.py")).write_text("""\
import joblib
from typing import List, Tuple

def predict_texts(texts: List[str], model_path: str = "models/baseline/model.pkl") -> Tuple[list, list]:
    pipe = joblib.load(model_path)
    preds = pipe.predict(texts)
    probs = pipe.predict_proba(texts)
    return preds.tolist(), probs.tolist()
""")

# ---------------- run_end_to_end.py ----------------
(Path(BASE / "run_end_to_end.py")).write_text("""\
import os
from pathlib import Path
from src.train_baseline import train_baseline
from src.evaluate import evaluate
from src.diagnostics import feature_importance, calibration_plot

BASE_DIR = Path(__file__).resolve().parent
DATA_PATH = BASE_DIR / "data" / "raw" / "reviews.csv"

if __name__ == "__main__":
    (BASE_DIR / "reports").mkdir(parents=True, exist_ok=True)
    train_df, val_df, test_df, val_pred, test_pred = train_baseline(str(DATA_PATH))
    # Validation
    evaluate(val_df['label'], val_pred,
             out_json=str(BASE_DIR/"reports"/"metrics_validation.json"),
             out_png=str(BASE_DIR/"reports"/"confusion_matrix_validation.png"))
    # Test
    evaluate(test_df['label'], test_pred,
             out_json=str(BASE_DIR/"reports"/"metrics.json"),
             out_png=str(BASE_DIR/"reports"/"confusion_matrix.png"))
    # Diagnostics
    feature_importance(model_path=str(BASE_DIR/"models"/"baseline"/"model.pkl"),
                       out_png=str(BASE_DIR/"reports"/"feature_importance.png"))
    calibration_plot(model_path=str(BASE_DIR/"models"/"baseline"/"model.pkl"),
                     X_texts=test_df['text'].tolist(),
                     y_true=test_df['label'].tolist(),
                     out_png=str(BASE_DIR/"reports"/"calibration_curve.png"))
    print("End-to-end baseline completed. Check reports/ and models/ folders.")
""")

print("Wrote source files under:", BASE / "src")
print("Runner script:", BASE / "run_end_to_end.py")
```

```python

from pathlib import Path

# ---------------- src/config.py ----------------
(Path(BASE / "src" / "config.py")).write_text("""\
RANDOM_STATE = 42
TEST_SIZE = 0.2
VAL_SIZE = 0.1
MAX_FEATURES = 100_000
NGRAM_RANGE = (1, 2)
LOGREG_C = 2.0
LOGREG_MAX_ITER = 1000
TOP_K_FEATURES = 20
N_BINS_CALIBRATION = 10
""")

# ---------------- src/preprocess.py ----------------
(Path(BASE / "src" / "preprocess.py")).write_text("""\
import re
import pandas as pd
from sklearn.model_selection import train_test_split
from .config import RANDOM_STATE, TEST_SIZE, VAL_SIZE

def clean_text(s: str) -> str:
    s = s.lower()
    s = re.sub(r"[^a-z0-9\\s']", " ", s)
    s = re.sub(r"\\s+", " ", s).strip()
    return s

def load_and_split(path: str, test_size: float = TEST_SIZE, val_size: float = VAL_SIZE):
    df = pd.read_csv(path)
    if 'text' not in df.columns or 'label' not in df.columns:
        raise ValueError("Input CSV must have columns: 'text' and 'label'.")
    df['text'] = df['text'].astype(str).apply(clean_text)
    train_df, test_df = train_test_split(
        df, test_size=test_size, stratify=df['label'], random_state=RANDOM_STATE
    )
    train_df, val_df = train_test_split(
        train_df, test_size=val_size, stratify=train_df['label'], random_state=RANDOM_STATE
    )
    return train_df.reset_index(drop=True), val_df.reset_index(drop=True), test_df.reset_index(drop=True)
""")

# ---------------- src/train_baseline.py (same train, but returns pipeline too) ----------------
(Path(BASE / "src" / "train_baseline.py")).write_text("""\
import json, os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from .preprocess import load_and_split
from .config import NGRAM_RANGE, MAX_FEATURES, LOGREG_C, LOGREG_MAX_ITER

def train_baseline(data_path: str,
                   model_out: str = "models/baseline/model.pkl",
                   report_out: str = "reports/metrics_baseline.json"):
    train_df, val_df, test_df = load_and_split(data_path)
    pipe = Pipeline([
        ("tfidf", TfidfVectorizer(ngram_range=NGRAM_RANGE, max_features=MAX_FEATURES)),
        ("clf", LogisticRegression(max_iter=LOGREG_MAX_ITER, C=LOGREG_C, class_weight="balanced"))
    ])
    pipe.fit(train_df['text'], train_df['label'])

    # Validate
    val_pred = pipe.predict(val_df['text'])
    metrics = {
        "val_accuracy": float(accuracy_score(val_df['label'], val_pred)),
        "val_f1": float(f1_score(val_df['label'], val_pred, average="binary")),
        "val_report": classification_report(val_df['label'], val_pred, output_dict=True)
    }

    # Save model & metrics
    import joblib
    os.makedirs(os.path.dirname(model_out), exist_ok=True)
    os.makedirs(os.path.dirname(report_out), exist_ok=True)
    joblib.dump(pipe, model_out)
    with open(report_out, "w") as f:
        json.dump(metrics, f, indent=2)

    # Test
    test_pred = pipe.predict(test_df['text'])
    test_metrics = {
        "test_accuracy": float(accuracy_score(test_df['label'], test_pred)),
        "test_f1": float(f1_score(test_df['label'], test_pred, average="binary")),
        "confusion_matrix": confusion_matrix(test_df['label'], test_pred).tolist()
    }
    os.makedirs("reports", exist_ok=True)
    with open("reports/metrics_baseline_test.json", "w") as f:
        json.dump(test_metrics, f, indent=2)

    return pipe, train_df, val_df, test_df, val_pred, test_pred
""")

# ---------------- src/metrics_and_plots.py (NEW: richer visualization + metrics) ----------------
(Path(BASE / "src" / "metrics_and_plots.py")).write_text("""\
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    roc_curve, auc, RocCurveDisplay,
    precision_recall_curve, average_precision_score,
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report, brier_score_loss, f1_score
)
from sklearn.calibration import calibration_curve
from .config import TOP_K_FEATURES, N_BINS_CALIBRATION

sns.set(style="whitegrid")

def save_json(d, path):
    with open(path, "w") as f:
        json.dump(d, f, indent=2)

def plot_confusion(y_true, y_pred, out_png, normalize=False, title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred, normalize='true' if normalize else None)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues", values_format=".2f" if normalize else None)
    plt.title(title)
    plt.tight_layout()
    plt.savefig(out_png, dpi=200, bbox_inches="tight")
    plt.close()

def plot_roc_pr(y_true, y_prob, out_roc, out_pr):
    # ROC
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = auc(fpr, tpr)
    RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc).plot()
    plt.title(f"ROC Curve (AUC={roc_auc:.3f})")
    plt.tight_layout()
    plt.savefig(out_roc, dpi=200, bbox_inches="tight")
    plt.close()
    # PR
    precision, recall, _ = precision_recall_curve(y_true, y_prob)
    ap = average_precision_score(y_true, y_prob)
    plt.figure()
    plt.plot(recall, precision, label=f"AP={ap:.3f}")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title("Precision-Recall Curve")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_pr, dpi=200, bbox_inches="tight")
    plt.close()
    return roc_auc, ap

def reliability_diagram(y_true, y_prob, out_png, n_bins=N_BINS_CALIBRATION):
    frac_pos, mean_pred = calibration_curve(y_true, y_prob, n_bins=n_bins)
    plt.figure(figsize=(6,6))
    plt.plot(mean_pred, frac_pos, "s-", label="Baseline")
    plt.plot([0,1],[0,1], "k--", label="Perfectly calibrated")
    plt.xlabel("Mean predicted probability")
    plt.ylabel("Fraction of positives")
    plt.title("Calibration Curve")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()

def ece_score(y_true, y_prob, n_bins=N_BINS_CALIBRATION):
    bins = np.linspace(0.0, 1.0, n_bins + 1)
    ece = 0.0
    for i in range(n_bins):
        lo, hi = bins[i], bins[i+1]
        mask = (y_prob >= lo) & (y_prob < hi)
        if mask.sum() == 0:
            continue
        bin_prob = y_prob[mask].mean()
        bin_acc  = y_true[mask].mean()
        ece += (mask.sum() / len(y_true)) * abs(bin_acc - bin_prob)
    return float(ece)

def prob_histograms(y_true, y_prob, out_png):
    plt.figure(figsize=(8,5))
    sns.histplot(y_prob[y_true==0], color="#d62728", label="Negative (true)", bins=20, stat="density", alpha=0.5)
    sns.histplot(y_prob[y_true==1], color="#2ca02c", label="Positive (true)", bins=20, stat="density", alpha=0.5)
    plt.xlabel("Predicted probability (positive class)")
    plt.title("Predicted Probability Distribution by True Class")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()

def threshold_search(y_true, y_prob, out_png):
    import numpy as np
    ts = np.linspace(0.05, 0.95, 19)
    scores = []
    for t in ts:
        preds = (y_prob >= t).astype(int)
        scores.append(f1_score(y_true, preds))
    plt.figure()
    plt.plot(ts, scores, "o-")
    plt.xlabel("Threshold"); plt.ylabel("F1 score")
    plt.title("F1 vs Threshold (Validation)")
    best_idx = int(np.argmax(scores))
    plt.axvline(ts[best_idx], color="k", linestyle="--", label=f"Best t={ts[best_idx]:.2f}, F1={scores[best_idx]:.3f}")
    plt.legend()
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()
    return float(ts[best_idx]), float(scores[best_idx])

def feature_importance(pipe, out_pos_png, out_neg_png, top_k=TOP_K_FEATURES):
    tfidf = pipe.named_steps["tfidf"]
    clf   = pipe.named_steps["clf"]
    import numpy as np
    vocab = np.array(tfidf.get_feature_names_out())
    coefs = clf.coef_[0]
    # Top positive (most indicative of positive), top negative
    pos_idx = np.argsort(coefs)[-top_k:][::-1]; neg_idx = np.argsort(coefs)[:top_k]
    pos_words, pos_vals = vocab[pos_idx], coefs[pos_idx]
    neg_words, neg_vals = vocab[neg_idx], coefs[neg_idx]
    # Positive
    plt.figure(figsize=(10,6))
    plt.barh(range(top_k), pos_vals[::-1], color="#2ca02c")
    plt.yticks(range(top_k), pos_words[::-1])
    plt.title("Top Positive TFâ€“IDF Features (LogReg Coefficients)")
    plt.gca().invert_yaxis(); plt.tight_layout()
    plt.savefig(out_pos_png, dpi=200); plt.close()
    # Negative
    plt.figure(figsize=(10,6))
    plt.barh(range(top_k), (-neg_vals)[::-1], color="#d62728")
    plt.yticks(range(top_k), neg_words[::-1])
    plt.title("Top Negative TFâ€“IDF Features (LogReg Coefficients)")
    plt.gca().invert_yaxis(); plt.tight_layout()
    plt.savefig(out_neg_png, dpi=200); plt.close()

def misclassification_table(df, y_true, y_prob, out_csv, k=15):
    import pandas as pd
    preds = (y_prob >= 0.5).astype(int)
    df_ = df.copy()
    df_["true"] = y_true
    df_["pred"] = preds
    df_["prob_pos"] = y_prob
    # false positives (pred=1, true=0), most confident
    fp = df_[(df_["true"]==0) & (df_["pred"]==1)].sort_values("prob_pos", ascending=False).head(k)
    # false negatives (pred=0, true=1), least confident
    fn = df_[(df_["true"]==1) & (df_["pred"]==0)].sort_values("prob_pos", ascending=True).head(k)
    out = pd.concat({"false_positives": fp, "false_negatives": fn}, axis=0)
    out.to_csv(out_csv, index=False)
    return out
""")

# ---------------- run_end_to_end.py (NEW orchestration) ----------------
(Path(BASE / "run_end_to_end.py")).write_text("""\
import os, json
from pathlib import Path
import numpy as np
from sklearn.metrics import classification_report
from src.train_baseline import train_baseline
from src.metrics_and_plots import (
    save_json, plot_confusion, plot_roc_pr, reliability_diagram,
    ece_score, prob_histograms, threshold_search, feature_importance,
    misclassification_table
)
from src.config import N_BINS_CALIBRATION

BASE_DIR = Path(__file__).resolve().parent
DATA_PATH = BASE_DIR / "data" / "raw" / "reviews.csv"
REPORTS = BASE_DIR / "reports"
REPORTS.mkdir(parents=True, exist_ok=True)

if __name__ == "__main__":
    pipe, train_df, val_df, test_df, val_pred, test_pred = train_baseline(str(DATA_PATH))
    # Probabilities
    val_prob = pipe.predict_proba(val_df['text'])[:, 1]
    test_prob = pipe.predict_proba(test_df['text'])[:, 1]

    # Core metrics (val & test)
    val_report = classification_report(val_df['label'], val_pred, output_dict=True)
    test_report = classification_report(test_df['label'], test_pred, output_dict=True)

    # ROC & PR curves (on test)
    roc_auc, ap = plot_roc_pr(test_df['label'].values, test_prob,
                              out_roc=str(REPORTS/"roc_curve.png"),
                              out_pr=str(REPORTS/"pr_curve.png"))
    # Calibration
    reliability_diagram(test_df['label'].values, test_prob, out_png=str(REPORTS/"calibration_curve.png"), n_bins=N_BINS_CALIBRATION)
    # ECE & Brier
    from sklearn.metrics import brier_score_loss
    ece = ece_score(test_df['label'].values, test_prob, n_bins=N_BINS_CALIBRATION)
    brier = brier_score_loss(test_df['label'].values, test_prob)

    # Probability histograms by true class
    prob_histograms(test_df['label'].values, test_prob, out_png=str(REPORTS/"probability_histograms.png"))

    # Confusion matrices: raw and normalized
    plot_confusion(test_df['label'].values, (test_prob>=0.5).astype(int), out_png=str(REPORTS/"confusion_matrix.png"), normalize=False)
    plot_confusion(test_df['label'].values, (test_prob>=0.5).astype(int), out_png=str(REPORTS/"confusion_matrix_normalized.png"), normalize=True, title="Confusion Matrix (Normalized)")

    # Threshold search on validation
    best_t, best_f1 = threshold_search(val_df['label'].values, val_prob, out_png=str(REPORTS/"f1_vs_threshold.png"))

    # Feature importance (pos & neg)
    feature_importance(pipe, out_pos_png=str(REPORTS/"feature_importance_positive.png"),
                       out_neg_png=str(REPORTS/"feature_importance_negative.png"))

    # Misclassification table (top errors)
    mis = misclassification_table(test_df[["text"]].copy(), test_df["label"].values, test_prob, out_csv=str(REPORTS/"top_misclassifications.csv"), k=15)

    # Save metrics JSON
    save_json({
        "validation": {"report": val_report},
        "test": {
            "report": test_report,
            "roc_auc": roc_auc,
            "average_precision": ap,
            "brier_score": brier,
            "ece": ece,
            "best_threshold": best_t,
            "best_f1_at_threshold": best_f1
        }
    }, str(REPORTS/"metrics_rich.json"))

    print("Done. Plots & metrics saved in:", REPORTS)
""")

print("Wrote improved source and runner to:", BASE)
```
